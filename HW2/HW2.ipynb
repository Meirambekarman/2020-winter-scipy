```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
data = pd.read_csv('company_sales_data.csv')
y = data['total_profit']
```

```python
fig = plt.figure(constrained_layout=True,figsize = (20,12))
gs = gridspec.GridSpec(2, 2)

ax = fig.add_subplot(gs[0, :])
y = data['total_profit']
x = data['month_number']

ax.set_title('total_profit by mounth')
ax.set_ylabel('Total profit')
ax.set_xlabel('Mounth')

plt.plot(x,y,
         color = 'r',linestyle = 'dashed',label = 'profit 1')
ax.legend(loc='upper left')
ax = fig.add_subplot(gs[1, 0])

j = 0
F = data.columns
j = 0
ax.set_xlabel('Mounth')
ax.set_title('all parameters by mounth')
for i in data:
    cs = F[j]
    j = j+1
    plt.plot(data[i],label = cs)
ax.legend(loc='upper left')

ax = fig.add_subplot(gs[1, 1])

x = data['month_number']
z = data['total_units']
ax.set_xlabel('Mounth')
width = 0.35
ax.bar(x-width/2,y,width,label = 'profit 1',align='center')
ax.bar(x+width/2,z,width,label = 'units',align='center')
ax.legend(loc='upper left')
```
