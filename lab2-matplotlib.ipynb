```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
data = pd.read_csv('company_sales_data.csv', sep=',')
data
```

```python
y = np.array(data['total_profit'])
x = np.array(data['month_number'])
print(x,y)
```

```python
plt.plot(x,y,marker='*', color='olive', linewidth=2, linestyle='dashed', label='Total')
plt.title("Total profin in month")
plt.xlabel("Month number")
plt.grid()
plt.ylabel("Total profit")
plt.xticks(x)
plt.legend()
```

```python
y1=np.array(data['facecream'])
y2=np.array(data['facewash'])
y3=np.array(data['toothpaste'])
y4=np.array(data['bathingsoap'])
y5=np.array(data['shampoo'])
y6=np.array(data['moisturizer'])
```

```python
plt.plot(x,y1, color='olive', linewidth=2, linestyle='dotted',marker='o', label='facecream')
plt.plot(x,y2, linewidth=2, linestyle='dashdot', marker="s", label='facewash')
plt.plot(x,y3, linewidth=2, linestyle='dashed', marker=">", label='toothpaste')
plt.plot(x,y4, linewidth=2, linestyle='solid', marker="*", label='bathingsoap')
plt.plot(x,y5, linewidth=2, linestyle=':', marker=".", label='shampoo')
plt.plot(x,y6, linewidth=2, linestyle='-.', marker=",", label='moisturizer')
plt.title("Total profin in month")
plt.xlabel("Month number")
plt.grid()
plt.ylabel("Total profit")
plt.xticks(x)
plt.legend(loc=(1.01, 0))
```

```python
fig, ax = plt.subplots()
width=0.35
rects1 = ax.bar(x - width/2, y1, width, label='facecream')
rects2 = ax.bar(x + width/2, y2, width, label='facewash')

ax.set_ylabel('Total profit')
ax.set_title('Total profit by facecream and facewash')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()
```

```python
import matplotlib.gridspec as gridspec

plt.figure(figsize=(10, 8))
G = gridspec.GridSpec(2, 2)

axes_1 = plt.subplot(G[0, :])
plt.xticks(())
plt.yticks(())
plt.plot(x,y,marker='', color='olive', linewidth=2, linestyle='dashed', label='Total')
plt.title("Total profit in month")
plt.xlabel("Month number")
plt.grid()
plt.ylabel("Total profit")
plt.xticks(x)
plt.legend(shadow = False, fontsize = 15)

axes_2 = plt.subplot(G[1, :-1])
plt.xticks(())
plt.yticks(())
plt.plot(x,y1, color='olive', linewidth=2, linestyle='dotted',marker='o', label='facecream')
plt.plot(x,y2, linewidth=2, linestyle='dashdot', marker="s", label='facewash')
plt.plot(x,y3, linewidth=2, linestyle='dashed', marker=">", label='toothpaste')
plt.plot(x,y4, linewidth=2, linestyle='solid', marker="*", label='bathingsoap')
plt.plot(x,y5, linewidth=2, linestyle=':', marker=".", label='shampoo')
plt.plot(x,y6, linewidth=2, linestyle='-.', marker=",", label='moisturizer')
plt.title("Total profit in month")
plt.xlabel("Month number")
plt.grid()
plt.ylabel("Total profit")
plt.xticks(x)
plt.legend(loc=(0.01, 0.78),ncol=2)

axes_3 = plt.subplot(G[1:, -1])
width=0.35
rects1 = plt.bar(x - width/2, y1, width, label='facecream')
rects2 = plt.bar(x + width/2, y2, width, label='facewash')

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('Total profit')
plt.title('Total profit by facecream and facewash')
plt.xticks(x)
plt.xlabel('Month number')
plt.legend(shadow = False, fontsize = 15)

plt.tight_layout()
plt.show()
```
