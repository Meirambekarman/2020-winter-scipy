```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
def csv_reader(file_obj):
    """
    Read a csv file
    """
    reader = csv.reader(file_obj)
    for row in reader:
        print(" ".join(row))
        
```

```python
csv_path = "company_sales_data.csv"
with open(csv_path, "r") as f_obj:
    csv_reader(f_obj)
```

```python
x = month_number=np.array(csv_file['month_number'])
y = np.array(csv_file['total_profit'])
y1 = facecream=np.array(csv_file['facecream'])
y2 = facewash=np.array(csv_file['facewash'])
y3 = toothpaste=np.array(csv_file['toothpaste'])
y4 = bathingsoap=np.array(csv_file['bathingsoap'])
y5 = shampoo=np.array(csv_file['shampoo'])
y6 = moisturizer=np.array(csv_file['moisturizer'])
#month_number=np.array(csv_file['month_number'])

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(month_number, facecream, marker='x', label = 'facecream')
ax.plot(month_number, facewash, marker='x', label = 'facewash')
ax.plot(month_number, toothpaste, marker='x', label = 'toothpaste')
ax.plot(month_number, bathingsoap, marker='x', label = 'bathingsoap')
ax.plot(month_number, shampoo, marker='x', label = 'shampoo')
ax.plot(month_number, moisturizer, marker='x', label = 'moisturizer')
ax.legend(bbox_to_anchor=(1.05, 1), loc = 'upper  right', borderaxespad=0.)
ax.set(xlabel='Month', ylabel='Good', title='')
#plt.
#a.legend(['df1','df2'])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
rects1 = ax.bar(month_number - 0.35/2, facewash, 0.35, label='facewash')
rects2 = ax.bar(month_number + 0.35/2, facecream, 0.35, label='facecream')
ax.set_ylabel('total profit')
ax.set_xlabel('month number')
ax.set_title('Scores by facewash and facecream')
ax.set_xticks(month_number)
ax.legend(bbox_to_anchor=(1, 1), loc = 'upper right', borderaxespad=0.)
```
