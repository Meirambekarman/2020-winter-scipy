```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
data = pd.read_csv('company_sales_data.csv')
data
```

```python
pltlegend=plt.plot(data['total_profit'], color='green',label="total profit", linestyle='--',marker='.',
                   markerfacecolor='blue', markersize=8)
plt.title("Matplotlib total_profit")
plt.grid()
plt.ylabel("Y label")
plt.xlabel("X label")
plt.legend(loc='upper left')
```

```python
data_month_number = data['month_number']

n_groups = len(data_month_number)
means_frank = (90, 55, 40, 65)
means_guido = (85, 62, 54, 20)

# create plot
# fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, data['toothpaste'], bar_width,
alpha=opacity,
color='b',
label='toothpaste')

rects2 = plt.bar(index + bar_width, data['bathingsoap'], bar_width,
alpha=opacity,
color='g',
label='bathingsoap')

plt.xticks(index + bar_width, data_month_number)
plt.legend()

plt.tight_layout()
plt.show()
```
