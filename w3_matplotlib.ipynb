```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('company_sales_data.csv')
df
```

```python
x = month_number = np.array(df['month_number'])
y = total_profit = np.array(df['total_profit'])
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(month_number, total_profit)
ax.set(xlabel='Months', ylabel='Total profits', title='')
```

```python
y1 = facecream = np.array(df['facecream'])
y2 = facewash = np.array(df['facewash'])
y3 = toothpaste = np.array(df['toothpaste'])
y4 = bathingsoap = np.array(df['bathingsoap'])
y5 = shampoo = np.array(df['shampoo'])
y6 = moisturizer = np.array(df['moisturizer'])

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(month_number, facecream, marker='o', label = 'facecream')
ax.plot(month_number, facewash, marker='o',label = 'facewash')
ax.plot(month_number, toothpaste, marker='o',label = 'toothpaste')
ax.plot(month_number, bathingsoap, marker='o',label = 'bathingsoap')
ax.plot(month_number, shampoo, marker='o',label = 'shampoo')
ax.plot(month_number, moisturizer, marker='o',label = 'moisturizer')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set(xlabel='Months', ylabel='Products', title='')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.bar(month_number - 0.35/2, facewash, 0.35, label='facewash')
ax.bar(month_number + 0.35/2, facecream, 0.35, label='facecream')
ax.set_ylabel('Products')
ax.set_xlabel('Months')
ax.set_title('Products by month')
ax.set_xticks(month_number)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
